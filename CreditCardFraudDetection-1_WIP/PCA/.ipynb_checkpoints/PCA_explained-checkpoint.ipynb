## Specific Columns

Lets investigate the
* Time Column
* Class Column
* Ammount Column


### Time Column

The time column has a very large range of values

```python
df['Time'].unique()
```

```python
df['Time'].unique().shape
```

This appears to be because the time is measured in seconds from a start time.  It would be more useful to convert this to a time of day, so lets convert it to hours and take the modulus with respect to 24 to get the hour of the day

```python
df["Time"] = df["Time"].apply(lambda x : (x / 3600) % 24)
```

Re-running the describe command for the *Time* column gives

```python
df['Time'].describe().T
```

which is as expected

### Class Column

Since the *Class* column is the label for the dataset, lets see how the labels are distributed

```python
df['Class'].value_counts()
```

Thus the dataset is VERY imbalanced, we will need to keep this in mind during later steps

### Amount Column

Lets see what the distribution of the ammounts is

```python
bins = 500
ax0 = plt.hist(df['Amount'], bins=bins)
# ax0.set_xlabel('Amount ($)')
```

This is an exponential decay with respect to ammount, which makes sense.

## Compare classes

We want to compare the classes with respect to counts and time


### Compare with respect to Counts

```python
df_fraud = df[df['Class'] == 1]
df_ok = df[df['Class'] == 0]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
bins = 30

ax1.hist(df_fraud['Amount'], bins=bins)
ax2.hist(df_ok['Amount'], bins=bins)

ax1.set_title('Fraud')
ax2.set_title('Normal')

ax1.set_xlabel('Amount ($)')
ax2.set_xlabel('Amount ($)')
ax1.set_ylabel('Number of Transactions')

ax2.set_ylim(0, 320)
plt.show()
```

Again, we get an exponential decay with respect to Amount for the number of transactions.  This makes sense as most transactions are for smaller amounts.

### Compare with respect to Times

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
bins = 30

ax1.hist(df_fraud['Time'], bins=bins)
ax2.hist(df_ok['Time'], bins=bins)

ax1.set_title('Fraud')
ax2.set_title('Normal')

ax1.set_xlabel('Time [h]')
ax2.set_xlabel('Time [h]')
ax1.set_ylabel('Number of Transactions')

ax2.set_yscale('log')
plt.show()
```

Here, we see that the fraudulent transactions are very random with respect to when they occur, while if we shift the normal transactions by 4 hours, we obtain a normal distribution.

```python
df1 = df0.copy()
df1["Time"] = df1["Time"].apply(lambda x : ((x / 3600) - 4) % 24)

df1_fraud = df1[df1['Class'] == 1]
df1_ok = df1[df1['Class'] == 0]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
bins = 30

ax1.hist(df1_fraud['Time'], bins=bins)
ax2.hist(df1_ok['Time'], bins=bins)

ax1.set_title('Fraud')
ax2.set_title('Normal')

ax1.set_xlabel('Time [h]')
ax2.set_xlabel('Time [h]')
ax1.set_ylabel('Number of Transactions')

ax2.set_yscale('log')
plt.show()
```

## Techniques looking to find anomolies by measuring reconstruction differences

First we need to generate datasets

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

and split them into test and training sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape
```

### PCA

First, we instantiate a PCA model and then call *fit* on it

```python
pca = PCA(n_components=10)

pca.fit(X_train)
```

#### Evaluate on Training Data

Let's evaluate the performance of our PCA model on the training data.  To do this we use our PCA model to transform and then untransform our training data

```python
X_train_trans = pca.transform(X_train)
X_train_invtrans = pca.inverse_transform(X_train_trans)

# Convert the inverse transformed traning data to a Pandas dataframe
X_train_invtrans = pd.DataFrame(X_train_invtrans, columns=X_train.columns, index=X_train.index)
X_train_invtrans.head()
```

Now, lets calculate the reconstruction scores and then convert them to a Pandas dataframe

```python
reconstruction_score = []
for idx in X_train.index:
    score = mean_squared_error(X_train.loc[idx], X_train_invtrans.loc[idx])
    reconstruction_score.append(score)

rc_scores = pd.DataFrame(reconstruction_score, index=X_train.index, columns=['reconstruction_score'])
```

Now, lets get the basic statistics for our reconstruction scores and then sort those scores

```python
rec_mean = rc_scores['reconstruction_score'].mean()
rec_median = rc_scores['reconstruction_score'].median()
rec_std = rc_scores['reconstruction_score'].std()

# Sorting reconstruction scores
rc_scores = rc_scores.sort_values(by='reconstruction_score', ascending=False)
```

##### Visualize

Now lets plot the reconstruction scores by the number of reconstruction scores in each of 60 buckets

```python
plt.figure(figsize=(10, 6))
rc_scores['reconstruction_score'].hist(bins=60)
plt.ylim(0, 100)
plt.title('PCA Reconstruction Score')
plt.xlabel('Reconstruction Score')
plt.ylabel('Number of samples')
```

##### Precision, Recall, F-1 Score

Finally, we want to get the precision and recall.  To do this, we hypothesize that the points with top scores will be the fraudulent transactions.  Thus, we need to get the top scores indicies of the top scores

```python
top_scores_idx = rc_scores[(rc_scores > (rec_median + 2*rec_std))].dropna().index
top_scores_idx.shape
```

as well as the indicies of the instances of fraud in the training dataset

```python
train_fraud_index = list(y_train[y_train == 1].index)
```

We also need a dataframe with the same indexing as our training data, but that assigns the entries with the highest reconstruction scores as fraud

```python
pred = pd.DataFrame(index=X_train.index)
pred['fraud'] = 0
for x in top_scores_idx:
    pred.loc[x, 'fraud'] = 1
```

This allows us to get the following recall, precision, and f-1 score information

```python
print(classification_report(y_train, pred['fraud']))
print('Rate of transations to investigate:', len(top_scores_idx) / len(X_train) * 100, '%')
```

##### More Precision, Recall, and F-1 Scores

For fun, lets do the following variations on calculating the top scores:
* 3 standard-deviations above the median
* 2 standard-deviations above the mean
* 3 standard-deviations above the mean

```python
top_scores_idx = rc_scores[(rc_scores > (rec_median + 3*rec_std))].dropna().index

train_fraud_index = list(y_train[y_train == 1].index)
pred = pd.DataFrame(index=X_train.index)
pred['fraud'] = 0
for x in top_scores_idx:
    pred.loc[x, 'fraud'] = 1

print(classification_report(y_train, pred['fraud']))
print('Rate of transations to investigate:', len(top_scores_idx) / len(X_train) * 100, '%')
print(top_scores_idx.shape)
```
